```python
#17. Replace Salary Values['<=50K','>50K'] With 0 and 1

df.columns
```

```python
df['income'].unique()
```

```python
df['income'].value_counts()
```

```python
sns.countplot('income',data=df)
```

```python
def salary_data(income):
    if income=='<=50k':
        return 0
    else:
        return 1
```

```python
df['encoded_income']=df['income'].apply(salary_data)
```

```python
df.head(1)
```

```python
df.replace(to_replace=['<=50k','>50k'],value=[0,1],inplace=True)
```

```python
df['income_numeric'] = df['income'].apply(lambda x: 1 if x == '>50K' else 0)
df.groupby('workclass')['income_numeric'].mean().sort_values(ascending=False)
```
